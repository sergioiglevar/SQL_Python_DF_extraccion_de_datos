##### Obtener el código, nombre y precio de los productos (estos dos últimos en el mismo campo)  <br> que están contenidos en los pedidos que ha tomado nota "Luis" o "María Luisa". <br> Ordena el listado de mayor a menor valor por fecha del pedido.

```python
connection = mysql.connector.connect(host=' ', database=' ', user=' ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query10 = """ SELECT P.Codigo, CONCAT(P.Nombre, ' - ', P.Precio) AS Nombre_Precio  , fecha FROM PEDIDO D 
JOIN consta C ON D.Numero = C.Numero_P JOIN PRODUCTO P ON C.Codigo_Pr = P.Codigo  
WHERE  D.DNI_EP IN (
SELECT DNI FROM EMPLEADO WHERE Nombre IN ('Luis Ramírez Pardo', 'María Luisa Galdón Ter') )
ORDER BY D.Fecha DESC;
      """
df10 = pd.read_sql(query10, con=connection)
df10
```

#### **Ejercicio 11**

##### Obtener por cada repartidor, su nombre, cantidad de pedidos y el tiempo <br> medio que tardan en entregar los pedidos una vez preparados. Ordenar el listado el tiempo medio que tardan en entregarlos.

```python
connection = mysql.connector.connect(host=' ', database=' ', user=' ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query11 = """ SELECT R.Nombre AS Repartidor, 
    COUNT(P.Numero) AS Cantidad_Pedidos, 
    AVG(TIMESTAMPDIFF(MINUTE, P.Hora_pre, P.Hora_rep)) AS Tiempo_Medio_Minutos  
FROM 
    REPARTIDOR R  
LEFT JOIN 
    PEDIDO P ON R.DNI = P.DNI_R  
GROUP BY 
    R.DNI, R.Nombre  
ORDER BY 
    Tiempo_Medio_Minutos;
      """
df11 = pd.read_sql(query11, con=connection)
df11
```

```python
connection = mysql.connector.connect(host=' ', database=' ', user=' ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query12 = """ SELECT P.Codigo, P.Nombre, P.Precio
FROM PRODUCTO P
WHERE
    P.Precio = (SELECT 
            MIN(Precio)
        FROM
            PRODUCTO)
        OR P.Precio = (SELECT 
            MAX(Precio)
        FROM
            PRODUCTO);
      """
df12 = pd.read_sql(query12, con=connection)
df12
```

#### **Ejercicio 13**

#####  Tener por cada producto , el nombre y código el número total de pedidos en los que se encuentra <br> teniendo en cuenta que el total de pedidos en los cuales se encuentre sea superior o igual a dos.<br> Ordena el listado de mayor a menor número de productos.

```python
connection = mysql.connector.connect(host=' ', database=' ', user=' ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query13 = """ SELECT P.Codigo,  P.Nombre, COUNT(C.Numero_P) AS Total_Pedidos 
FROM PRODUCTO P 
JOIN  consta C ON P.Codigo = C.Codigo_Pr 
GROUP BY  P.Codigo, P.Nombre 
HAVING COUNT(C.Numero_P) >= 2 
ORDER BY Total_Pedidos DESC;
      """
df13 = pd.read_sql(query13, con=connection)
df13
```

####  **Ejercicio 15**

##### Obtener el nombre del producto que es menú junto con el código de losproductos <br> que lo componen en aquellos pedidos del mes de septiembre de 2020.

<br>

```python
connection = mysql.connector.connect(host=' ', database=' ', user=' ', password='  ', connection_timeout=180000)
cursor=connection.cursor()
query15 = """ SELECT P.Nombre AS Nombre_Menu, EC.Codigo_P_compuesto AS Codigo_Producto FROM PEDIDO PD  
JOIN consta C ON PD.Numero = C.Numero_P  
JOIN PRODUCTO P ON C.Codigo_Pr = P.Codigo  
JOIN esta_compuesto EC ON P.Codigo = EC.Codigo_P  
WHERE P.Nombre LIKE 'Menú%'   AND PD.Fecha >= '2020-09-01' AND PD.Fecha < '2020-10-01';       """
df15 = pd.read_sql(query15, con=connection)
df15
```
